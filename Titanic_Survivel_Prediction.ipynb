```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/tested.csv")
```

```python
df.head()
```

```python
#Mapping text data to binary value
df['Sex']=df['Sex'].map({'male':1,'female':0}).astype(int)
```

```python
df
```

```python
X=df[['Pclass','Sex','Age','Fare']]
X
```

```python
Y=df.Survived
Y
```

```python
# checking null value and replace it
X.isnull().sum()
```

```python
Y.isnull().sum()
```

```python
X.Age=X.Age.fillna(X.Age.mean())
X.Fare=X.Age.fillna(X.Fare.mean())
```

```python
X.isnull().sum()
```

```python
#splitting dataset into training and testing part
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
# training
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,Y_train)
```

```python
# Predicting whether person survived or not
pclassNo=int(input("enter person's Pclass number: "))
gender=int(input("enter person male:{1} or female:{0}: "))
age=int(input("enter person's age: "))
fare=float(input("enter person's fare: "))
person=[[pclassNo,gender,age,fare]]
result=model.predict(person)
print(result)

if result==1:
    print("might be survived")
else:
    print("might not be survived")
```
